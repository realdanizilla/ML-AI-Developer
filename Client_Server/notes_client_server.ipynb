# Client and Server

## Introdução 

Aborda a parte de redes

Cliente pede ou enviar informações para servidor. Exemplos
- browser
- celular quando manda msg de whatsapp

servidor fica ligado 24hs, as vezes é físico mas no mais das vezes é cloud, servidor web alugado AWS, GCP, AZURE

comunicação acontece nesse modelo cliente-servidor

servidor cria um loop infinito que espera o cliente se conectar com ele. Através de um fluxo de informação o servidor vai saber pra onde enviar / receber informações

cliente sempre começa a conversa e envia informações

servidor ta sempre em loop infinito ouvindo (listening) e dá conta de vários clientes (paralelismo, assíncrono), então pode ser usado para esperar uma conexão ou para realizar uma conexão. Se ele realiza uma conexão ele está sendo um client

desenho básico:



usamos cliente-servidor para Modelos de ML e IA que são servidos através de uma API Rest

Protocolo padrão é o TCP - sempre garante que todos os pacotes enviados serão recebidos, porém é lenta - se a informação é sensível e vou precisar dela usar TCP

O protocolo UDP permite que um ou mais pacotes não cheguem e ele consegue inferir o que tava nos pacotes faltantes - utilizado em jogos

quando pensamos em colocar nosso modelo de ML em produção temos que pensar na velocidade necessária para performar bem, pensa no anti-fraude da mastercard com milhoes de pessoas passando cartões em todos os lugares ao mesmo tempo

## Aula 3 - HTTP e HTTPS

Um protocolo tal qual o TCP e o UDP

HTTPS é criptografada pelo cliente e decodificada pelo serviços

Preciso saber usar os verbos

métodos que podemos usar
- get - requisição de informação para o servidor (ex: página do site)
- post - enviando informação para o servidor (ex: preencher um formulário)
- put - atualizar alguma coisa
- delete - deletar algo no banco de dados


possíveis respostas do servidor:
- código 200 = sucesso, entendido!
- código classe 300 = transferencia de requisição
- código 400 - erro do cliente, inclusive o 404
- código 500 - cliente mandou correto mas servidor nao conseguiu devolver



## Aula 4 - API REST

Endpoint - ponto de conexão entre a sua aplicação e o que está fora dela - é uma URL ilustrativa para o que eu quero (ex: www.url.com/consult/v1/enviar)

- endpoint predict (Seria um post, ele recebe uma imagem e devolve uma previsão do que é a imagem)

tem mais controle e mais estrutura que HTTP

Fast Api é um framework que ajuda a desenvolver API, não é um servidor, muito utilizada por causa de requisição assíncrona

tem que configurar cada endpoint do lado do servidor e do lado do cliente para que o servidor possa fazer o que o cliente quer (get, post, put, delete)

temos alguns tipos de conexões:
- socket TCP - cano com coisa dentro, poe string de um lado sai string de outro
- socket UDP - cano com coisa dentro, não tem garantia de entrega dos dados
- HTTP - um pouco mais de estrutura como o verbo
- API REST - mais controle e estrutura, defino verbo e endpoint

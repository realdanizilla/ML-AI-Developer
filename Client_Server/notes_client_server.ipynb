## Aula 3 - HTTP e HTTPS

Um protocolo tal qual o TCP e o UDP

HTTPS  criptografada pelo cliente e decodificada pelo servios

Preciso saber usar os verbos

mtodos que podemos usar
- get - requisio de informao para o servidor (ex: pgina do site)
- post - enviando informao para o servidor (ex: preencher um formulrio)
- put - atualizar alguma coisa
- delete - deletar algo no banco de dados


possveis respostas do servidor:
- cdigo 200 = sucesso, entendido!
- cdigo classe 300 = transferencia de requisio
- cdigo 400 - erro do cliente, inclusive o 404
- cdigo 500 - cliente mandou correto mas servidor nao conseguiu devolver



## Aula 4 - API REST

Endpoint - ponto de conexo entre a sua aplicao e o que est fora dela -  uma URL ilustrativa para o que eu quero (ex: www.url.com/consult/v1/enviar)

- endpoint predict (Seria um post, ele recebe uma imagem e devolve uma previso do que  a imagem)

tem mais controle e mais estrutura que HTTP

Fast Api  um framework que ajuda a desenvolver API, no  um servidor, muito utilizada por causa de requisio assncrona

tem que configurar cada endpoint do lado do servidor e do lado do cliente para que o servidor possa fazer o que o cliente quer (get, post, put, delete)

temos alguns tipos de conexes:
- socket TCP - cano com coisa dentro, poe string de um lado sai string de outro
- socket UDP - cano com coisa dentro, no tem garantia de entrega dos dados
- HTTP - um pouco mais de estrutura como o verbo
- API REST - mais controle e estrutura, defino verbo e endpoint
